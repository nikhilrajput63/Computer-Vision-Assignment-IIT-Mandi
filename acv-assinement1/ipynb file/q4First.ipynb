```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
from torch.utils.data import DataLoader, random_split
import time
import copy

torch.manual_seed(42)
np.random.seed(42)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

batch_size = 128
num_epochs = 30
learning_rate = 0.001
weight_decay = 1e-4

mean = [0.4914, 0.4822, 0.4465]
std = [0.2470, 0.2435, 0.2616]

train_transform = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
    transforms.ToTensor(),
    transforms.Normalize(mean, std)
])
test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean, std)
])

full_train_dataset = torchvision.datasets.CIFAR10(
    root='./data', train=True, download=True, transform=train_transform)

train_size = int(0.8 * len(full_train_dataset))
val_size = len(full_train_dataset) - train_size
train_subset, val_subset = random_split(full_train_dataset, [train_size, val_size])

train_loader = DataLoader(train_subset, batch_size=batch_size, shuffle=True, num_workers=2, pin_memory=True)

val_dataset = torchvision.datasets.CIFAR10(
    root='./data', train=True, download=True, transform=test_transform)
val_indices = val_subset.indices
val_dataset_with_test_transform = torch.utils.data.Subset(val_dataset, val_indices)
val_loader = DataLoader(val_dataset_with_test_transform, batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)

test_dataset = torchvision.datasets.CIFAR10(
    root='./data', train=False, download=True, transform=test_transform)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')

class CNN(nn.Module):
    def __init__(self, filter_size=3, use_batch_norm=True, activation_fn='relu'):
        super(CNN, self).__init__()
        padding = filter_size // 2
        if activation_fn == 'relu':
            self.activation = nn.ReLU()
        elif activation_fn == 'leaky_relu':
            self.activation = nn.LeakyReLU(0.1)
        elif activation_fn == 'elu':
            self.activation = nn.ELU()
        else:
            self.activation = nn.ReLU()
        self.conv1 = nn.Conv2d(3, 64, kernel_size=filter_size, padding=padding)
        self.bn1 = nn.BatchNorm2d(64) if use_batch_norm else nn.Identity()
        self.conv2 = nn.Conv2d(64, 128, kernel_size=filter_size, padding=padding)
        self.bn2 = nn.BatchNorm2d(128) if use_batch_norm else nn.Identity()
        self.conv3 = nn.Conv2d(128, 256, kernel_size=filter_size, padding=padding)
        self.bn3 = nn.BatchNorm2d(256) if use_batch_norm else nn.Identity()
        self.pool = nn.MaxPool2d(2, 2)
        self.fc = nn.Linear(256 * 4 * 4, 10)
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        x = self.pool(self.activation(self.bn1(self.conv1(x))))
        x = self.pool(self.activation(self.bn2(self.conv2(x))))
        x = self.pool(self.activation(self.bn3(self.conv3(x))))
        x = x.view(-1, 256 * 4 * 4)
        x = self.dropout(x)
        x = self.fc(x)
        return x

def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs, scheduler=None):
    best_acc = 0.0
    best_model_wts = copy.deepcopy(model.state_dict())
    best_epoch = 0
    train_losses = []
    val_losses = []
    val_accs = []
    for epoch in range(num_epochs):
        start_time = time.time()
        model.train()
        running_loss = 0.0
        for i, (images, labels) in enumerate(train_loader):
            images = images.to(device)
            labels = labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * images.size(0)
        epoch_train_loss = running_loss / len(train_loader.dataset)
        train_losses.append(epoch_train_loss)
        model.eval()
        running_loss = 0.0
        correct = 0
        total = 0
        with torch.no_grad():
            for images, labels in val_loader:
                images = images.to(device)
                labels = labels.to(device)
                outputs = model(images)
                loss = criterion(outputs, labels)
                running_loss += loss.item() * images.size(0)
                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()
        epoch_val_loss = running_loss / len(val_loader.dataset)
        val_losses.append(epoch_val_loss)
        epoch_val_acc = correct / total
        val_accs.append(epoch_val_acc)
        if scheduler:
            if isinstance(scheduler, optim.lr_scheduler.ReduceLROnPlateau):
                scheduler.step(epoch_val_loss)
            else:
                scheduler.step()
        epoch_time = time.time() - start_time
        print(f'Epoch [{epoch+1}/{num_epochs}], '
              f'Train Loss: {epoch_train_loss:.4f}, '
              f'Val Loss: {epoch_val_loss:.4f}, '
              f'Val Acc: {epoch_val_acc:.4f}, '
              f'Time: {epoch_time:.2f}s')
        if epoch_val_acc > best_acc:
            best_acc = epoch_val_acc
            best_model_wts = copy.deepcopy(model.state_dict())
            best_epoch = epoch
    model.load_state_dict(best_model_wts)
    return model, train_losses, val_losses, val_accs, best_epoch

def evaluate_model(model, test_loader):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in test_loader:
            images = images.to(device)
            labels = labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    test_acc = correct / total
    return test_acc

def plot_curves(train_losses, val_losses, val_accs, best_epoch):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.axvline(x=best_epoch, color='r', linestyle='--', label=f'Best Model (Epoch {best_epoch+1})')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Training and Validation Loss')
    plt.grid(True)
    plt.subplot(1, 2, 2)
    plt.plot(val_accs, label='Validation Accuracy')
    plt.axvline(x=best_epoch, color='r', linestyle='--', label=f'Best Model (Epoch {best_epoch+1})')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.title('Validation Accuracy')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def run_experiment(filter_size, use_batch_norm, activation_fn, learning_rate, weight_decay):
    print(f"\nExperiment: Filter Size={filter_size}, Batch Norm={use_batch_norm}, "
          f"Activation={activation_fn}, LR={learning_rate}, Weight Decay={weight_decay}")
    model = CNN(filter_size=filter_size, use_batch_norm=use_batch_norm, activation_fn=activation_fn).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=weight_decay)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=3, factor=0.5)
    model, train_losses, val_losses, val_accs, best_epoch = train_model(
        model, train_loader, val_loader, criterion, optimizer, num_epochs, scheduler)
    test_acc = evaluate_model(model, test_loader)
    print(f"Test Accuracy: {test_acc:.4f}")
    plot_curves(train_losses, val_losses, val_accs, best_epoch)
    return model, test_acc, best_epoch

experiments = [
    (3, True, 'relu', 0.001, 1e-4),
    (5, True, 'relu', 0.001, 1e-4),
    (3, False, 'relu', 0.001, 1e-4),
    (3, True, 'leaky_relu', 0.001, 1e-4),
    (3, True, 'elu', 0.001, 1e-4),
    (3, True, 'relu', 0.0005, 1e-4),
    (3, True, 'relu', 0.001, 1e-3)
]

results = []

for params in experiments:
    model, test_acc, best_epoch = run_experiment(*params)
    results.append((params, test_acc, best_epoch))

print("\n--- Experiment Results ---")
for i, ((filter_size, use_batch_norm, activation_fn, lr, wd), test_acc, best_epoch) in enumerate(results):
    print(f"Experiment {i+1}:")
    print(f"  Filter Size: {filter_size}")
    print(f"  Batch Normalization: {use_batch_norm}")
    print(f"  Activation Function: {activation_fn}")
    print(f"  Learning Rate: {lr}")
    print(f"  Weight Decay: {wd}")
    print(f"  Best Epoch: {best_epoch+1}")
    print(f"  Test Accuracy: {test_acc:.4f}")
    print()

best_idx = max(range(len(results)), key=lambda i: results[i][1])
best_params, best_test_acc, best_epoch = results[best_idx]

print("\n--- Best Model ---")
print(f"Experiment {best_idx+1}:")
print(f"  Filter Size: {best_params[0]}")
print(f"  Batch Normalization: {best_params[1]}")
print(f"  Activation Function: {best_params[2]}")
print(f"  Learning Rate: {best_params[3]}")
print(f"  Weight Decay: {best_params[4]}")
print(f"  Best Epoch: {best_epoch+1}")
print(f"  Test Accuracy: {best_test_acc:.4f}")

print("\n--- Training Final Model with Best Parameters ---")
filter_size, use_batch_norm, activation_fn, learning_rate, weight_decay = best_params

final_model = CNN(filter_size=filter_size, use_batch_norm=use_batch_norm, activation_fn=activation_fn).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(final_model.parameters(), lr=learning_rate, weight_decay=weight_decay)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=3, factor=0.5, verbose=True)
final_model, train_losses, val_losses, val_accs, best_epoch = train_model(
    final_model, train_loader, val_loader, criterion, optimizer, num_epochs, scheduler)
test_acc = evaluate_model(final_model, test_loader)
print(f"Final Model Test Accuracy: {test_acc:.4f}")

plt.figure(figsize=(15, 5))
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.plot(val_accs, label='Validation Accuracy')
plt.axvline(x=best_epoch, color='r', linestyle='--', label=f'Best Model (Epoch {best_epoch+1})')
plt.xlabel('Epoch')
plt.ylabel('Loss / Accuracy')
plt.legend()
plt.title('Training Curves for Best Model')
plt.grid(True)
plt.tight_layout()
plt.show()

torch.save(final_model.state_dict(), 'cifar10_cnn_best.pth')
print("Best model saved to 'cifar10_cnn_best.pth'")
```

**Training Final Model with Best Parameters**

```python
# Final model training with best parameters
print("\n--- Training Final Model with Best Parameters ---")
filter_size = 3  
use_batch_norm = True  
activation_fn = 'leaky_relu'  
learning_rate = 0.001  
weight_decay = 1e-4  

final_model = CNN(filter_size=filter_size, use_batch_norm=use_batch_norm, activation_fn=activation_fn).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(final_model.parameters(), lr=learning_rate, weight_decay=weight_decay)

scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=3, factor=0.5)

# Train and evaluate
final_model, train_losses, val_losses, val_accs, best_epoch = train_model(
    final_model, train_loader, val_loader, criterion, optimizer, num_epochs, scheduler)

# Evaluate on test data
test_acc = evaluate_model(final_model, test_loader)
print(f"Final Model Test Accuracy: {test_acc:.4f}")

plt.figure(figsize=(15, 5))

# Plot all curves in same figure
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.plot(val_accs, label='Validation Accuracy')
plt.axvline(x=best_epoch, color='r', linestyle='--', label=f'Best Model (Epoch {best_epoch+1})')
plt.xlabel('Epoch')
plt.ylabel('Loss / Accuracy')
plt.legend()
plt.title('Training Curves for Best Model')
plt.grid(True)

plt.tight_layout()
plt.show()

# Save the model
torch.save(final_model.state_dict(), 'cifar10_cnn_best.pth')
print("Best model saved to 'cifar10_cnn_best.pth'")
```
